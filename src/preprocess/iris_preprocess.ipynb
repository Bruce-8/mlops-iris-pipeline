# Preprocessing of the Iris Dataset

## 1. Import Libraries

```python
import pandas as pd
import os
```

## 2. Load Dataset

```python
df = pd.read_csv('../../data/iris.csv')
```

## 3. Exploratory Data Analysis

```python
# Display the first 5 rows of the dataset.
print(df.head())
```

```python
# Display some statistical data about the dataset.
print(df.describe())
```

```python
# Display the information about the dataset.
print(df.info())
```

## 4. Handle Missing Values

```python
#Check for missing values
print(df.isnull().sum())
```

## 5. Remove Duplicate Records

```python
# Identify duplicate records
print(f"Number of duplicate records: {df.duplicated().sum()}")

# Remove duplicate records
df = df.drop_duplicates()

# Check for duplicate records again
print(f"Number of duplicate records: {df.duplicated().sum()}")
```

## 6. Data Types and Conversion

```python
# Display the data types of the dataset
print(df.dtypes)
```

## 7. Encode Categorical Variables

```python
df["species"] = df["species"].astype("category").cat.codes

# Display the info of the dataset
print(df.info())
```

## 8. Feature Selection

```python
# Compute correlation matrix for numeric columns
corr_matrix = df.corr(numeric_only=True)

# Inspect correlation of each feature with the target 'species'
corr_with_target = corr_matrix['species'].drop('species').sort_values(ascending=False)
print("Correlation of features with target 'species':")
print(corr_with_target)

# Select features whose absolute correlation with the target is above a threshold
threshold = 0.3
selected_features = corr_with_target[abs(corr_with_target) > threshold].index.tolist()
print("\nSelected features based on correlation threshold:", selected_features)
```

## 9. Exporting Cleaned Dataset

```python
# Ensure output directory exists (from notebook at src/preprocess/)
output_dir = "../processed_data"
os.makedirs(output_dir, exist_ok=True)

# Export cleaned dataset to CSV
output_path = os.path.join(output_dir, "iris_cleaned.csv")
df.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to: {output_path}")
```
